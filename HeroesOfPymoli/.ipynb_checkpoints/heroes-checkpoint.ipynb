# Option 1: Heroes of Pymoli: Main script with the analysis and results

```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load 
games = "Resources/Lessons_04-Pandas_Homework_Instructions_HeroesOfPymoli_Resources_purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
games_df = pd.read_csv(games)
games_df.head()
```

# Player Count
 
Display the total number of players


```python
# Variable calculations
players=len(games_df["SN"].unique())

#DataFrame with results
total_players=pd.DataFrame({"Total Players" :[players]})

#Display the total number of players
total_players.head()
```

# Purchasing Analysis (Total)
 
Run basic calculations to obtain number of unique items, average price, etc.

Create a summary data frame to hold the results

Optional: give the displayed data cleaner formatting

Display the summary data frame

```python
#variables calculations
unique_items=len(games_df["Item ID"].unique())
average_price=games_df["Price"].mean()
num_purchases=len(games_df["Purchase ID"].unique())
total_revenue=games_df["Price"].sum()

#Dataframe with results
summary_df=pd.DataFrame({"Number of Unique Items" :[unique_items],
                         "Average Price":[average_price],
                         "Number of Purchases": [num_purchases],
                         "Total Revenue": [total_revenue]})
 
# Use Map to format columns
summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,}".format)

#Display Reults
summary_df.head()                         
```

# Gender Demographics

Percentage and Count of Male Players

Percentage and Count of Female Players

Percentage and Count of Other / Non-Disclosed


```python
#Variables calculations
males_df = games_df.loc[games_df["Gender"] == "Male", :]
total_males=len(males_df["SN"].unique())
total_males

females_df = games_df.loc[games_df["Gender"] == "Female", :]
total_females=len(females_df["SN"].unique())
total_females

other_df = games_df.loc[games_df["Gender"] == "Other / Non-Disclosed", :]
total_others=len(other_df["SN"].unique())
total_others

p_males=total_males/players*100
p_females=total_females/players*100
p_others=total_others/players*100

#Dataframe with results
total_gender_df=pd.DataFrame({"Gender": ["Male", "Female", "Other / Non-Disclosed"],
                           "Total Count": [total_males,total_females,total_others],
                          "Percentage of Players": [p_males,p_females, p_others]})

                           
# Use Map to format columns
total_gender_df["Percentage of Players"] = total_gender_df["Percentage of Players"].map("{:.2f}%".format)

#Display Reults
total_gender_df.style.hide_index()
```

# Purchasing Analysis (Gender)

Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

Create a summary data frame to hold the results

Optional: give the displayed data cleaner formatting

Display the summary data frame

```python
#variables calculations
gender_group= games_df.groupby (['Gender'])
purchases=gender_group["Purchase ID"].count()
average_price= gender_group["Price"].mean()
purchase_value= gender_group["Price"].sum()
average_purchase_person=purchase_value/[total_females,total_males,total_others]


#Dataframe with results
gender_summary_df = pd.DataFrame({"Purchase Count": purchases,
                                    "Average Purchase Price": average_price,
                                 "Total Purchase Value": purchase_value,
                                "Avg Total Purchase per Person":average_purchase_person})

# Use Map to format columns
gender_summary_df["Average Purchase Price"] = gender_summary_df["Average Purchase Price"].map("${:.2f}".format)
gender_summary_df["Total Purchase Value"] = gender_summary_df["Total Purchase Value"].map("${:,.2f}".format)
gender_summary_df["Avg Total Purchase per Person"] = gender_summary_df["Avg Total Purchase per Person"].map("${:.2f}".format)

#Display Reults
gender_summary_df.head()                         
```

From the dataset, we have that most of the players are males and a small proportion is females or Other / Non-Disclosed. Therefore, we can see that most of the purchases come from male players.
However, the average value purchase per person does not differ significantly by gender. This is that players on average purchase the same amount of items with similar prices independently if they are males, females or others.
In this sense, we can expect that if there were more female or Other / Non-Disclosed players, they will contribute with similar per person revenues than men.

# Age Demographics

The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)

Purchase Count

Average Purchase Price

Total Purchase Value

Average Purchase Total per Person by Age Group


```python
#Generate a copy od the original dataframe to eliminate SN duplicates
games_copy=games_df.copy()
clean_df=games_copy.drop_duplicates(subset=["SN"])
clean_df

# Create bins and group names
bins=[0, 9,14,19,24,29,34,39,100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Add column with Age Ranges
clean_df["Age Ranges"] = pd.cut(clean_df["Age"], bins, labels=group_names, include_lowest=True)

#Groupby and variables calculations

age_group=clean_df.groupby(["Age Ranges"])
Total_Count=age_group["Age Ranges"].count()
p_age=Total_Count/players*100

#DataFrame with results
age_summary_df = pd.DataFrame({"Total Count": Total_Count,
                              "Percentage of Players": p_age})

# Use Map to format columns
age_summary_df["Percentage of Players"] = age_summary_df["Percentage of Players"] .map("{:.2f}%".format)

#Display Reults
age_summary_df.head(8)

```

# Top Spenders

Identify the the top 5 spenders in the game by total purchase value, then list (in a table):

SN

Purchase Count

Average Purchase Price

Total Purchase Value

```python
#Group by SN 
SN_group=games_df.groupby(["SN"])

#variables calculations 
purchases_sn=SN_group["Purchase ID"].count()
average_price_sn= SN_group["Price"].mean()
purchase_value_sn= SN_group["Price"].sum()


#DataFrame with results
sn_summary_df = pd.DataFrame({"Purchase Count": purchases_sn,
                                    "Average Purchase Price": average_price_sn,
                                 "Total Purchase Value": purchase_value_sn})

sn_sorted_df = sn_summary_df.sort_values("Total Purchase Value", ascending=False)

# Use Map to format columns
sn_sorted_df["Average Purchase Price"] = sn_sorted_df["Average Purchase Price"].map("${:.2f}".format)
sn_sorted_df["Total Purchase Value"] = sn_sorted_df["Total Purchase Value"].map("${:,.2f}".format)

#Display Reults
sn_sorted_df.head() 
```

# Most Popular Items

Identify the 5 most popular items by purchase count, then list (in a table):

Item ID

Item Name

Purchase Count

Item Price

Total Purchase Value

```python
#Retrieve the Item ID, Item Name, and Item Price columns
items_df= games_df.loc[:,["Item ID","Item Name", "Price"]]

#Group by Item ID and Item Name
item_group=items_df.groupby(["Item ID", "Item Name"])

#variables calculations
purchases_item=item_group["Item ID"].count()
average_price_item= item_group["Price"].mean()
purchase_value_item= purchases_item *average_price_item


#DataFrame with results
item_summary_df = pd.DataFrame({"Purchase Count": purchases_item,
                                    "Item Price": average_price_item,
                                 "Total Purchase Value": purchase_value_item})

#sorted by Purchase Count

item_sorted_df = item_summary_df.sort_values("Purchase Count", ascending=False)

# Use Map to format columns
item_sorted_df["Item Price"] = item_sorted_df["Item Price"].map("${:.2f}".format)
item_sorted_df["Total Purchase Value"] = item_sorted_df["Total Purchase Value"].map("${:,.2f}".format)

#Display Reults
item_sorted_df.head(10) 
```

# Most Profitable Items

Identify the 5 most profitable items by total purchase value, then list (in a table):

Item ID

```python
#sorted by Total Purchase Value
item_sorted2_df = item_summary_df.sort_values("Total Purchase Value", ascending=False)

# Use Map to format columns
item_sorted2_df["Item Price"] = item_sorted2_df["Item Price"].map("${:.2f}".format)
item_sorted2_df["Total Purchase Value"] = item_sorted2_df["Total Purchase Value"].map("${:,.2f}".format)

#Display Reults
item_sorted2_df.head(10) 
```

Analyzing the purchases by item, we observe that the sold items top ten have at least 8 purchases. If we consider the total purchase value, we get that the top ten has a value greater than $25. 

Considering both variables, the two most popular (more number of purchases) and most profitable items are "Final Critic" and "Oathbreaker, Last Hope of the Breaking Storm", with more than 10 purchases each and a total purchase value greater than $50; followed by "Fiery Glass Crusader", "Persuasion", "Extraction, Quickblade Of Trembling Hands",  "Nirvana" and "Singed Scalpel".
